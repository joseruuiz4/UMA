```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np



df = pd.read_csv("./bmw.csv")
```

**Ejercicio 1:**
Representa la funci√≥n f(x) = sin(x)+cos(x) en el intervalo [0,2*pi].


```python
x = np.linspace(0, 2*np.pi, 100)

y = np.sin(x) + np.cos(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gr√°fica de f(x) = sin(x) + cos(x)')
plt.grid(True)
plt.show()
```

**Ejercicio 2:**
Representa las funciones f(x) = sin(x), g(x)=log(1+x) en el intervalo [0,2*pi], tomando 100 puntos igualmente espaciados en dicho intervalo.


```python
y_sin = np.sin(x)
y_log = np.log(1 + x)

plt.plot(x, y_sin, label='f(x) = sin(x)')
plt.plot(x, y_log, label='g(x) = log(1+x)')
plt.xlabel('x')
plt.ylabel('f(x), g(x)')
plt.title('Gr√°ficas de f(x) = sin(x) y g(x) = log(1+x)')
plt.legend()
plt.grid(True)
plt.show()
```

**Ejercicio 6:**
Considera la siguiente funci√≥n, que calcula valores enteros entre 0 y 255 para dibujar el fractal de Mandelbrot, siendo a y b n√∫meros en coma flotante que representan la parte real y la parte imaginaria de un n√∫mero complejo, respectivamente.






```python
min_real = -2
max_real = 1
min_imaginaria = -1
max_imaginaria = 1

mapa_calor = np.zeros((400, 600))

MAXIMO_ITERACIONES = 80

def mandelbrot(a, b):
    c = complex(a, b)
    z = 0
    n = 0
    while abs(z) <= 2 and n < MAXIMO_ITERACIONES:
        z = z * z + c
        n += 1
    color_pixel = 255 - int(n * 255 / MAXIMO_ITERACIONES)
    return color_pixel


for y in range(400):
    for x in range(600):
        real = min_real + (max_real - min_real) * x / (600 - 1)
        imaginaria = min_imaginaria + (max_imaginaria - min_imaginaria) * y / (400 - 1)
        mapa_calor[y, x] = mandelbrot(real, imaginaria)


plt.imshow(mapa_calor, cmap='hot', extent=[min_real, max_real, min_imaginaria, max_imaginaria])
plt.colorbar(label='Iteraciones')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginaria')
plt.title('Mapa de calor del fractal de Mandelbrot')
plt.show()
```
