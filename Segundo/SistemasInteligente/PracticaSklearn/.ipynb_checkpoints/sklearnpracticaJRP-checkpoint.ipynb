```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import datasets
df = pd.read_csv('./winequality-red.csv',sep= ';')
#Se cambia el tipo de la columna quality a float
df['quality'] = df['quality'].astype(float)
print(df)

X = df.drop(["quality"], axis = 1)
y = df["quality"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)

# #Clasica los datos con un clasificador Decision Tree:
from sklearn.tree import DecisionTreeClassifier, plot_tree


# Grafica la curva de validacin
#CURVA DE VALIDACION PARA ELEGIR EL ALPHA
from sklearn.model_selection import validation_curve

# Rango de valores para ccp_alpha que queremos probar
param_range = np.linspace(0, 0.1, 10)


# Calcula la curva de validacin para ccp_alpha utilizando la precisin como mtrica
train_scores, test_scores = validation_curve(
    DecisionTreeClassifier(), X, y, param_name="ccp_alpha", param_range=param_range,
    scoring="accuracy", cv=5
)

# Calcula las medias y desviaciones estndar de los scores
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)


plt.figure(figsize=(10, 6))
plt.title("Curva de Validacin para ccp_alpha")
plt.xlabel("$\\alpha$")
plt.ylabel("Precisin")
plt.ylim(0.0, 1.1)
plt.semilogx(param_range, train_scores_mean, label="Training score", color="darkorange", lw=2)
plt.fill_between(param_range, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.2, color="darkorange")
plt.semilogx(param_range, test_scores_mean, label="Cross-validation score", color="navy", lw=2)
plt.fill_between(param_range, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.2, color="navy")
plt.legend(loc="best")
plt.show

dtc = DecisionTreeClassifier()
dtc.get_params()
dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_test)
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred)) 

print(classification_report(y_test,y_pred))
features = pd.DataFrame(dtc.feature_importances_,index=X.columns)
print(features)


dtc2 = DecisionTreeClassifier(criterion="entropy",ccp_alpha=0.04)
dtc2.fit(X_train,y_train)
y_pred2 = dtc2.predict(X_test)
print(confusion_matrix(y_test,y_pred)) 
print(classification_report(y_test,y_pred))
features2 = pd.DataFrame(dtc2.feature_importances_,index=X.columns)
print(features2)


dtc3 = DecisionTreeClassifier(criterion="entropy",splitter="random")
dtc3.fit(X_train,y_train)
y_pred2 = dtc3.predict(X_test)
print(confusion_matrix(y_test,y_pred)) 
print(classification_report(y_test,y_pred))
features3 = pd.DataFrame(dtc2.feature_importances_,index=X.columns)
print(features3)


dtc4 = DecisionTreeClassifier(criterion="log_loss",ccp_alpha=0.04)
dtc2.fit(X_train,y_train)
y_pred2 = dtc2.predict(X_test)
print(confusion_matrix(y_test,y_pred)) 
print(classification_report(y_test,y_pred))
features2 = pd.DataFrame(dtc2.feature_importances_,index=X.columns)
print(features2)


from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# Define el objeto KFold con 5 divisiones
kf = KFold(n_splits=5, shuffle=True, random_state=42)


# Listas para almacenar las mtricas de cada modelo en cada divisin
metrics_dtc = {'accuracy': [], 'precision': [], 'recall': [], 'f1_score': []}
metrics_dtc2 = {'accuracy': [], 'precision': [], 'recall': [], 'f1_score': []}
metrics_dtc3 = {'accuracy': [], 'precision': [], 'recall': [], 'f1_score': []}
metrics_dtc4 = {'accuracy': [], 'precision': [], 'recall': [], 'f1_score': []}


# Realiza la validacin cruzada para el primer DecisionTreeClassifier
for train_index, test_index in kf.split(X):
    X_train_fold, X_test_fold = X.iloc[train_index], X.iloc[test_index]
    y_train_fold, y_test_fold = y.iloc[train_index], y.iloc[test_index]
    dtc.fit(X_train_fold, y_train_fold)
    y_pred_fold = dtc.predict(X_test_fold)
    metrics_dtc['accuracy'].append(accuracy_score(y_test_fold, y_pred_fold,))
    metrics_dtc['precision'].append(precision_score(y_test_fold, y_pred_fold,average='macro',zero_division=1))
    metrics_dtc['recall'].append(recall_score(y_test_fold, y_pred_fold,average='macro'))
    metrics_dtc['f1_score'].append(f1_score(y_test_fold, y_pred_fold,average='macro'))


# Realiza la validacin cruzada para el segundo DecisionTreeClassifier
for train_index, test_index in kf.split(X):
    X_train_fold, X_test_fold = X.iloc[train_index], X.iloc[test_index]
    y_train_fold, y_test_fold = y.iloc[train_index], y.iloc[test_index]
    dtc2.fit(X_train_fold, y_train_fold)
    y_pred_fold = dtc2.predict(X_test_fold)
    metrics_dtc2['accuracy'].append(accuracy_score(y_test_fold, y_pred_fold))
    metrics_dtc2['precision'].append(precision_score(y_test_fold, y_pred_fold,average='macro',zero_division=1))
    metrics_dtc2['recall'].append(recall_score(y_test_fold, y_pred_fold,average='macro'))
    metrics_dtc2['f1_score'].append(f1_score(y_test_fold, y_pred_fold,average='macro'))


# Realiza la validacin cruzada para el tercer DecisionTreeClassifier
for train_index, test_index in kf.split(X):
    X_train_fold, X_test_fold = X.iloc[train_index], X.iloc[test_index]
    y_train_fold, y_test_fold = y.iloc[train_index], y.iloc[test_index]
    dtc3.fit(X_train_fold, y_train_fold)
    y_pred_fold = dtc3.predict(X_test_fold)
    metrics_dtc3['accuracy'].append(accuracy_score(y_test_fold, y_pred_fold,))
    metrics_dtc3['precision'].append(precision_score(y_test_fold, y_pred_fold,average='macro',zero_division=1))
    metrics_dtc3['recall'].append(recall_score(y_test_fold, y_pred_fold,average='macro'))
    metrics_dtc3['f1_score'].append(f1_score(y_test_fold, y_pred_fold,average='macro'))


# Realiza la validacin cruzada para el cuarto DecisionTreeClassifier
for train_index, test_index in kf.split(X):
    X_train_fold, X_test_fold = X.iloc[train_index], X.iloc[test_index]
    y_train_fold, y_test_fold = y.iloc[train_index], y.iloc[test_index]
    dtc4.fit(X_train_fold, y_train_fold)
    y_pred_fold = dtc4.predict(X_test_fold)
    metrics_dtc4['accuracy'].append(accuracy_score(y_test_fold, y_pred_fold,))
    metrics_dtc4['precision'].append(precision_score(y_test_fold, y_pred_fold,average='macro',zero_division=1))
    metrics_dtc4['recall'].append(recall_score(y_test_fold, y_pred_fold,average='macro'))
    metrics_dtc4['f1_score'].append(f1_score(y_test_fold, y_pred_fold,average='macro'))


# Calcula las mtricas promedio para cada modelo
avg_metrics_dtc = {metric: sum(values) / len(values) for metric, values in metrics_dtc.items()}
avg_metrics_dtc2 = {metric: sum(values) / len(values) for metric, values in metrics_dtc2.items()}
avg_metrics_dtc3 = {metric: sum(values) / len(values) for metric, values in metrics_dtc3.items()}
avg_metrics_dtc4 = {metric: sum(values) / len(values) for metric, values in metrics_dtc4.items()}


# Imprime las mtricas promedio para cada modelo
print("Mtricas para DecisionTreeClassifier con criterion = 'Giny' y cc_aplha = 0:\n")
print(avg_metrics_dtc)
print("\nMtricas para DecisionTreeClassifier con criterion = 'entropy' y ccp_alpha = 0.04:\n")
print(avg_metrics_dtc2)
print("\nMtricas para DecisionTreeClassifier con criterion = 'entropy',splitter = 'random'\n")
print(avg_metrics_dtc3)
print("\nMtricas para DecisionTreeClassifier con criterion = 'log_loss',ccp_alpha = 0.04:\n")
print(avg_metrics_dtc4)


plt.figure(figsize=(20,10))  # Ajusta el tamao del grfico
class_names = [str(class_name) for class_name in y.unique()]
plot_tree(dtc3, feature_names=X.columns, class_names= class_names, filled=True)
plt.show()



```

```python
from sklearn import metrics
from sklearn import neural_network
import warnings
warnings.filterwarnings('ignore')


clf = neural_network.MLPClassifier(hidden_layer_sizes= (10,10,10),max_iter=5000)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test); 
acur = metrics.classification_report(y_test,y_pred)
cm = metrics.confusion_matrix(y_test,y_pred)
print("Acurracy: ", acur)
print ("confusion matrix: \n",cm)
print("------------------------\n")


clf2 = neural_network.MLPClassifier(activation="logistic",solver="lbfgs")
clf2.fit(X_train,y_train)
y_pred2 = clf2.predict(X_test); 
acur2 = metrics.classification_report(y_test,y_pred)
cm2 = metrics.confusion_matrix(y_test,y_pred)
print("Acurracy: ", acur2)
print ("confusion matrix: \n",cm2)
print("------------------------\n")


clf3 = neural_network.MLPClassifier(learning_rate="adaptive", solver="adam")
clf3.fit(X_train,y_train)
y_pred3 = clf3.predict(X_test); 
acur3 = metrics.classification_report(y_test,y_pred)
cm3 = metrics.confusion_matrix(y_test,y_pred)
print("Acurracy: ", acur3)
print ("confusion matrix: \n",cm3)
print("------------------------\n")


clf4 = neural_network.MLPClassifier(alpha=0.03,learning_rate="constant")
clf4.fit(X_train,y_train)
y_pred4 = clf4.predict(X_test); 
acur4 = metrics.classification_report(y_test,y_pred)
cm4 = metrics.confusion_matrix(y_test,y_pred)
print("Acurracy: ", acur4)
print ("confusion matrix: \n",cm4)
print("------------------------\n")



from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Define el objeto KFold con 5 divisiones
kf = KFold(n_splits=5, shuffle=True, random_state=42)


# Listas para almacenar las mtricas de cada modelo en cada divisin
metrics_mlp1 = {'accuracy': [], 'precision': [], 'recall': [], 'f1_score': []}
metrics_mlp2 = {'accuracy': [], 'precision': [], 'recall': [], 'f1_score': []}
metrics_mlp3 = {'accuracy': [], 'precision': [], 'recall': [], 'f1_score': []}
metrics_mlp4 = {'accuracy': [], 'precision': [], 'recall': [], 'f1_score': []}


# Configuraciones de MLPClassifier
clf1 = neural_network.MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=5000)
clf2 = neural_network.MLPClassifier(activation="logistic", solver="lbfgs")
clf3 = neural_network.MLPClassifier(learning_rate="adaptive", solver="adam")
clf4 = neural_network.MLPClassifier(alpha=0.03, learning_rate="constant")


# Funcin para evaluar un modelo con KFold
def evaluate_model(clf, metrics_dict):
    for train_index, test_index in kf.split(X_scaled):
        X_train_fold, X_test_fold = X_scaled[train_index], X_scaled[test_index]
        y_train_fold, y_test_fold = y[train_index], y[test_index]
        clf.fit(X_train_fold, y_train_fold)
        y_pred_fold = clf.predict(X_test_fold)
        metrics_dict['accuracy'].append(accuracy_score(y_test_fold, y_pred_fold))
        metrics_dict['precision'].append(precision_score(y_test_fold, y_pred_fold, average='macro', zero_division=1))
        metrics_dict['recall'].append(recall_score(y_test_fold, y_pred_fold, average='macro'))
        metrics_dict['f1_score'].append(f1_score(y_test_fold, y_pred_fold, average='macro'))


# Evaluar cada modelo
evaluate_model(clf1, metrics_mlp1)
evaluate_model(clf2, metrics_mlp2)
evaluate_model(clf3, metrics_mlp3)
evaluate_model(clf4, metrics_mlp4)


# Calcula las mtricas promedio para cada modelo
avg_metrics_mlp1 = {metric: sum(values) / len(values) for metric, values in metrics_mlp1.items()}
avg_metrics_mlp2 = {metric: sum(values) / len(values) for metric, values in metrics_mlp2.items()}
avg_metrics_mlp3 = {metric: sum(values) / len(values) for metric, values in metrics_mlp3.items()}
avg_metrics_mlp4 = {metric: sum(values) / len(values) for metric, values in metrics_mlp4.items()}


# Imprime las mtricas promedio para cada modelo
print("Mtricas para MLPClassifier Modelo 1(hidden_layer_sizes=(10,10,10), max_iter=5000):\n")
print(avg_metrics_mlp1)
print("\nMtricas para MLPClassifier Modelo 2(activation='logistic', solver='lbfgs'):\n")
print(avg_metrics_mlp2)
print("\nMtricas para MLPClassifier Modelo 3(activation='logistic', solver='adam'):\n")
print(avg_metrics_mlp3)
print("\nMtricas para MLPClassifier Modelo 4(alpha=0.03, learning_rate='constant'):\n")
print(avg_metrics_mlp4)
```

### Curva roc para el perceptrn multicapa

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc


# 1. Convertir las etiquetas de clase a un formato binario
y_test_bin = label_binarize(y_test, classes=np.unique(y))


# 2. Entrenar el clasificador y predecir probabilidades
clf3.fit(X_train, y_train)
y_prob = clf3.predict_proba(X_test)


# 3. Calcular tasas de verdaderos positivos (TPR) y falsos positivos (FPR)
fpr, tpr, _ = roc_curve(y_test_bin.ravel(), y_prob.ravel())


# 4. Calcular el rea bajo la curva ROC (AUC)
roc_auc = auc(fpr, tpr)


# Traza la curva ROC
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC para el MLPClassifier Modelo3')
plt.legend(loc="lower right")
plt.show()
```
